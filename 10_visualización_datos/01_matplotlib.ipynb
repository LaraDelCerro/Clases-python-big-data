## 1.Gráfico de líneas(Line Plot)
**Caso de uso**: Mostrar tendencias y cambios a lo largo del tiempo.    
**Data Model**: Dos listas de datos (Eje x y Eje y independiente)


```python
import matplotlib.pyplot as plt
import pandas as pd

# Dataset: Ventas de una tienda por mes (listas nativas de Python)
#meses = ['Enero', 'Febrero' 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']

ventas = [10,15,12,20,25,22] #en miles de euros

plt.figure(figsize= (10,6))#tamaño del lienzo en pulgadas (ancho x alto)
plt.plot(meses,ventas, linestyle = ':', color ='blue', marker = 's')
plt.xlabel('Mes', color = 'blue')
plt.ylabel('Ventas (miles de €)', color = 'blue')
plt.grid()#marca la cuadrícula por debajo 
plt.title('Tendencia de ventas (Enero - Junio)', color = 'purple')
plt.show()
```

## 2. Gráfico de Barras Verticales (Vertical Bar Chart)

**Casos de uso**: Comparar cantidades entre las diferentes categorías individuales.                     
**Data Model**: Dos listas separadas (Categorías en eje x y Cantidades en eje y)

```python
frutas = ['Manzanas', 'Plátanos', 'Peras', 'Fresas', 'Mangos']
cantidad = [120, 85, 150, 240, 40] #udes disponibles

plt.figure(figsize=(10,6))
plt.bar(frutas, cantidad, color = 'violet', alpha = 0.8, width = 0.4, edgecolor= 'black', linewidth = 2, hatch = 'x') 
#alpha mide la intensidad del color
#width es el ancho de las barras
#edgecolor es el color del borde de las barras y linewidth es su grosor
plt.xlabel('Frutas', color='blue')
plt.ylabel('Cantidad (unidades)', color = 'blue')
plt.grid()#marca la cuadrícula por debajo 
plt.title('Inventario de cantidad de frutas', color = 'purple')
plt.show()
```

# Gráficos de Barras Horizontales (Horizontal Bar Chart)

**Casos de uso**: Compara categorías a partir de un listado estructurado de registros.     
**Data Model**: Una lista de diccionarios (simulando una BB.DD.), usando pandas.

```python
datos_lenguajes = [
    {"lenguaje": "Python", "votos": 35},
    {"lenguaje": "JavaScript", "votos": 28},
    {"lenguaje": "C++", "votos": 12},
    {"lenguaje": "Java", "votos": 18},
    {"lenguaje": "SQL", "votos": 22}
]

df_lenguajes = pd.DataFrame(datos_lenguajes)
lenguajes = df_lenguajes['lenguaje']
votos = df_lenguajes['votos']
plt.figure(figsize=(10,6))
#para pintar la gráfica extraemos directamente del df
plt.barh(lenguajes,votos)
plt.title('Lenguajes de programación favoritos')
plt.xlabel('Número de votos')
plt.ylabel('Lenguaje')
```
