# Datos 2 y 3: Rendimiento por tecnología
datos_tech = {
    "Tecnologia": ["Python", "SQL", "Node.js", "React", "Angular"],
    "Horas_Practica_Semanal": [12, 6, 8, 10, 5],
    "Satisfaccion_Alumnos": [4.8, 4.1, 4.4, 4.7, 3.9]
}

## Misión:

Gráfico de Líneas (Evolución):
Muestra el crecimiento de las matriculas a lo largo de los meses.
Requisitos: La línea debe ser punteada (linestyle='--'), de color azul y tener un marcador circular (marker='o') en cada punto.

Gráfico de Barras Verticales (Carga de trabajo):
Compara las Horas_Practica_Semanal requeridas para cada Tecnologia.
Requisitos: Todas las barras deben ser de color naranja con un borde negro (edgecolor='black') y un grosor de línea de 1.5.

Gráfico de Barras Horizontales (Valoración):
Muestra la Satisfaccion_Alumnos obtenida por cada Tecnologia.
Requisitos: Pon un color distinto para cada barra,


```python
# Datos 1: Evolución temporal
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
matriculas = [15, 22, 20, 28, 35, 40]

# Datos 2 y 3: Rendimiento por tecnología
datos_tech = {
    "Tecnologia": ["Python", "SQL", "Node.js", "React", "Angular"],
    "Horas_Practica_Semanal": [12, 6, 8, 10, 5],
    "Satisfaccion_Alumnos": [4.8, 4.1, 4.4, 4.7, 3.9]
}
```

```python
import matplotlib.pyplot as plt
import pandas as pd


plt.figure(figsize= (10,6))#tamaño del lienzo en pulgadas (ancho x alto)
plt.plot(meses, matriculas, linestyle = '--', color ='blue', marker = 'o')
plt.xlabel('Mes', color = 'blue')
plt.ylabel('Matriculas', color = 'blue')
plt.grid()#marca la cuadrícula por debajo 
plt.title('Tendencia de matrículas mensuales', color = 'purple')
plt.show()
```

## Gráfico de barras verticales

```python
df_datos_tech = pd.DataFrame(datos_tech)
horas_semanales = df_datos_tech['Horas_Practica_Semanal']
tecnologia = df_datos_tech['Tecnologia']
plt.figure(figsize=(10,6))
plt.bar(tecnologia,horas_semanales, color = 'orange', edgecolor= 'black', linewidth = 1.5) 
plt.title('Horas de práctica semanales por tecnología')
plt.xlabel('Horas Semanales')
plt.ylabel('Tecnologia')
plt.show()
```

## Gráfico de barras horizontales

```python
valoracion = df_datos_tech['Satisfaccion_Alumnos']
plt.figure(figsize=(10,6))
colores = ['Darkblue', 'Gray', 'Green', 'lightblue', 'purple']#array de 5 colores
plt.barh(tecnologia,valoracion, color = colores, edgecolor='black')
plt.title('Valoración para cada tecnología')
plt.xlabel('Valoración')
plt.ylabel('Tecnologia')
plt.show()
```
