# Selección y filtrado de datos con PANDAS

Seleccionar filas y columnas específicas es la operación más frecuente en Pandas. Hay tres mecanismos principales: acceso directo por nombre, `loc` (por etiqueta) e `iloc` (por posición entera). Entender cuándo usar cada uno evita errores habituales.

## Acceso directo por nombre de serie

```python
print(df)
print('')
#si queremos la una serie (columna):
print(df['edad'])
print('')
#si queremos sacar varias series:
print(df[['nombre', 'salario']])
```

## Acceso por Loc - selección por etiqueta

loc['filas, columnas'] usa los nombres del índice y de la columnas. **A diferencia de slicing de python** si tiene los limite incluidos

```python
 #fila por índice

print(df.loc[0]) #sería como devolver un diccionario con os datos del registro posición 0 
print('-'*30)
print(df.loc[1:3])#incluye el 3. Devuelve los 3 primeros registros. 
print('-'*30)
print(df.loc[1:6:2])
print('-'*30)
print(df.loc[6:1:-1])
```

 El siguiente punto interesante sería juntar la posibilidad de hacer una combinación de ambos. Es decir, filtrar nombre y salario del registro 3 al 5. 

```python
print(df.loc[3:5, ['nombre', 'salario']])
```

## set_index() y idx_loc -> recogemos el valor por índices numéricos

```python
#Crea un nuevo df donde los índices es la serie de nombres
df_idx = df.set_index('nombre')#convierte el nombre en el índice
print(df_idx)
print('')
print('### Datos de Eva ###')
print('')
print(df_idx.loc['Eva'])# saca los datos de Eva
```

 # Loc y df tienen un problema respecto al slicing de Python: no tienen posiciones negativas. 
 Para ello usamos ILOC que se basa en el slicing de Python, es como si trabajáramos con Python sin la librería pandas

```python
print(df.iloc[0:2]) #es equivalente al slicing de python df[0:2], no saca el último valor. Saco los dos primeros registros 
print('')
#iloc me permite índices negativos, como el slicing
print(df.iloc[-2:])#saca desde las posiciones -2 hasta el final, es decir, los índices 5 (equivalentea a -1) y 4(equivalente a -2)
print('')
#iloc me permite trabajar con filas y columnas
print(df.iloc[0:3, 0:2]) #saco los tres primeros registros y para cada uno las columnas 0 y 1 (nombre y departamento)
```

| | `loc` | `iloc` |
|-|-------|--------|
| Basado en | Etiquetas (nombres) | Posiciones enteras |
| Extremo final | **Incluido** | **Excluido** |
| Uso ideal | Índice significativo | Índice numérico o acceso posicional |

## Filtro con condiciones booleanas

Por ejemplo sacar los trabajadores que tenga un salario mayor que 30000. O un salario incluido en un rango 

```python
# lo primero sería la condición, es decir filtrar el df por la condición
# resultado = df['salario'] > 30000
# print(df[resultado])
print(df[df['salario']>30000]) #equivalente a las dos líneas anteriores


## en df and = &, or = | 
rango = (df['salario'] >= 2000) & (df['salario'] <= 4000)
print(df[rango])
```

### filtrar por una lista de valores 

```python
filtro = (df['departamento']== 'IT') | (df['departamento']== 'RRHH')
print(df[filtro])
print('')
# otra opción: isin, filtra por lista de valores
filtro2=df['departamento'].isin(['IT', 'RRHH'])
print(df[filtro2])
```

## Between: rango inclusivo

```python
rango_edades = df['edad'].between(25,35) # 25 y 35 incluidos
print(df[rango_edades])
```

## QUERY() -filtrado con expresión de tipo sql
Es la alternativa más legible para filtros complejos, inspirada en SQL

```python
# Equivale a df[(df["salario"] > 40000) & (df["departamento"] != "RRHH")]

resultado = df.query(" salario > 40000 and departamento !='RRHH' ")
print(resultado)
print('')
#Nos permite usar variable
limite = 50000
resultado2 = df.query("salario >= @limite")
print(resultado2)
print('')
dpt = 'RRHH'
resultado3 = df.query("departamento == @dpt")
print(resultado3)
```

```python
#Buscar aquellos datos de los trabajadores que se apellidan Pérez
resultado2 = df.query('nombre.str.contains("pérez", case = False)') #case= False no tiene en cuenta las mayúsculas/minúsculas
print(resultado2)
```

---
**Resumen:**
- `df["col"]` → Series de una columna
- `df[["col1", "col2"]]` → DataFrame de varias columnas
- `df.loc[fila, col]` → por etiqueta (extremo incluido)
- `df.iloc[fila, col]` → por posición entera (extremo excluido)
- `df[condición]` → filtro booleano
- `df.query("expr")` → filtro en texto estilo SQL
