```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "nombre":      ["Ana", "Luis", "Eva", "Carlos", "Marta", "Pedro", "Laura", "Tomás"],
    "departamento":["IT", "Ventas", "IT", "RRHH", "Ventas", "IT", "RRHH", "Ventas"],
    "edad":        [23, None, 28, 45, 22, 38, 35, 27],
    "salario":     [28000, 42000, None, 61000, 24000, 55000, 48000, 31000],
    "activo":      [True, True, False, True, True, True, False, True],
    "fecha_alta":  pd.to_datetime(["2022-01-10", "2019-06-15", "2021-03-20",
                                   "2015-11-01", "2023-02-28", "2017-08-12",
                                   "2020-04-05", "2021-09-18"])
})


print(df)
print('')
print(df.info())
print(df.describe())
```

# Tenemos exploración de columnas por categoría: value_counts, unique, nunique

```python
## value_counts: repetición de cada valor

print(df['departamento'].value_counts())
print('')
print(df['departamento'].value_counts(normalize=True)) # % de cada departamento. En IT trabaja el 37.5% de la plantilla
```

```python
# lista de valores únicos por departamento
print(df['departamento'].unique())
print('')
# nunique: número de valores únicos por columna
print(df.nunique())
```

## Valores nulos: primera detección

```python
print(df.isnull()) #imprime True en los nulo
#print(df.isna()) hace lo mismo

#TOTAL DE NULOS
print('TOTAL DE NULOS')
print(df.isnull().sum())

#¿Cuántos nulos hay en el campo salario?
print(f'Nulos en salario: ',df['salario'].isnull().sum())
```

---
**Resumen — checklist de exploración inicial:**
1. `df.shape` → dimensiones
2. `df.head()` / `df.tail()` → ver los datos
3. `df.info()` → tipos y nulos de un vistazo
4. `df.describe()` → estadísticas numéricas
5. `df[col].value_counts()` → distribución de categóricas
6. `df.isnull().sum()` → conteo de nulos
