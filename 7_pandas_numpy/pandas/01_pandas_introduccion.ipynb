# Introducción a Pandas

Pandas es la librería de referencia para análisis de datos tabulares en Python. Se construye sobre NumPy y añade dos estructuras fundamentales: **Series** (columna de datos) y **DataFrame** (tabla completa). La diferencia clave frente a NumPy es que los datos tienen **etiquetas** (índices y nombres de columna), lo que hace el código más legible y los datos más manejables

## Serie
Una serie es un array de 1D, es decir, un vector. En un df se corresponde con las columnas. Las filas son los registros

```python
import pandas as pd

serie = pd.Series([1,2.5,3,4,5,6])
print(serie)
print('')
print(serie[0])#accedemos a la serie como si fuera una lista
print('*'*15)
#pordemos ponerle un índice personalizado a mi serie
notas = pd.Series([1,2,3,4,5,6], 
                index = ['Ana','María', 'Pepe', 'Joaquín', 'Juan', 'Miguel'],
                name = 'nota_final'
                )

print(notas)
print('')
print(notas['Ana'])#accedemos a la serie como si fuera un diccionario

print('*'*15)
serie_dic = pd.Series({'manzanas': 1.20, 'peras': 0.90, "naranja": 0.75})
print(serie_dic)
print(serie_dic['peras'])
```

# DataFrame - una tabla con índices de fila y de columna

DataFrame es un conjunto de series(vectores) que comparte el mismo índice

```python
# Desde un diccionario de listas: claves → nombres de columna
df = pd.DataFrame({
    "nombre": ["Ana", "Luis", "Eva", "Carlos", "Marta"],
    "edad":   [23, 31, 28, 45, 22],
    "ciudad": ["Madrid", "Barcelona", "Sevilla", "Madrid", "Valencia"],
    "salario": [28000, 42000, 35000, 61000, 24000]
})

print(df)
```

```python
registros = [
    {"producto": "Laptop",  "precio": 999,  "stock": 15},
    {"producto": "Ratón",   "precio": 29,   "stock": 82},
    {"producto": "Monitor", "precio": 349,  "stock": 7},
]

df2 = pd.DataFrame(registros)
print(df2)
```

## Atributos fundamentales de un DataFrame


```python
print(f'shape(dimensión)', df.shape)
print(f'shape(dimensión)', df2.shape)

print('-'*30)

print('columnas', df.columns.tolist())
print('columnas', df2.columns.tolist())

print('-' * 30)

print('indices', df.index.tolist())

print('-' * 30)
print(df2.dtypes)
```

```python
# podemos ver solo las primeras 5 filas del df
print(df.head())
print('-'*40)
# podemos ver solo las últimas 2 filas
print(df.tail(2))

# ambas fcones se puden parametrizar con la cantidad que queramos, si no ponemos nada, por defecto saca 5 registros.
```

```python
# una muestra aleatoria de los datos
print (df.sample(3))
```
