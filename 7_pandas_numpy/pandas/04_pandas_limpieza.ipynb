```python
import pandas as pd
```

```python
# Dataset con problemas típicos
df = pd.DataFrame({
    "ID":           [1, 2, 3, 4, 4, 5, 6],
    "Nombre":       ["Ana", "Luis", "Eva", "Carlos", "Carlos", None, "Marta"],
    "Departamento": ["IT", "ventas", "it", "rrhh", "RRHH", "Ventas", "IT "],  # inconsistencias
    "Salario":      [28000, None, 35000, 61000, 61000, 48000, 24000],
    "Edad":         [23, 31, None, 45, 45, 38, 22],
})
print(df)
```

# Limpieza de datos

En proyectos reales, rara vez los datos llegan de forma correcta o perfectas condiciones. Esta es una de las fases más importantes del trabajo, detectar y tratar los datos inconsistentes, valores nulos, duplicados, errores, etc. 

## Valores nulos: isnull, dropna, fillna

```python
# Detectar nulos
print('Mapa de nulos')
print(df.isnull())
print('')

# Quiero detectar los nulos por columna. Más cómodo para visualizar si tengo valores nulos
print(df.isnull().sum())
```

```python
# eliminar filas que tengan algún nulo

df_sin_nulos = df.dropna() #elimina todos los registros que tengan algún valor nulo
print(df_sin_nulos)
```

```python
#eliminar sólo algunas columnas específicas

df_sin_nulos2 = df.dropna(subset=['Salario', 'Edad']) #elimina los registros que tengan salario o edad nulos
print(df_sin_nulos2)
```

```python
# fillna: rellena nulos con un valor concreto

df_relleno = df.copy() #Hago una copia y la llamo de_relleno
#Rellenar con un valor fijo
df_relleno['Nombre'] = df_relleno['Nombre'].fillna('Desconocido')

#Rellenar la mediana del salario en aquellos campos que sean nulos --> elegimos media o mediana en función de la dispersión de los datos. Mucha dispersión, es más fiable la mediana
mediana_salario = df_relleno['Salario'].median()
df_relleno['Salario'] = df_relleno['Salario'].fillna(mediana_salario)


#Rellenar con la media de la edad aquellos campos nulos
media_edad = df_relleno['Edad'].mean()
df_relleno['Edad'] = df_relleno['Edad'].fillna(media_edad)

print(df_relleno)
```

##  DUPLICADOS: duplicated(), drop_duplicates()

```python
# detectar filas duplicadas
duplicado = df_relleno.duplicated()
print(df[duplicado])
print('')

#eliminar duplicados(mantiene el primer registro por defecto y elimina el resto de duplicados)
df_resultante = df_relleno.drop_duplicates()
print(df_resultante)
```

## Renombrar columnas e índices de nuestro dataframe

```python
# cambiar el nombre de las columnas a minúsculas
# df2 = df.copy()
# df2.columns = df2.columns.str.lower()
# print(df2)
# print('#'*40)

df_rename = df_resultante.rename(columns={
    "ID": 'id',
    "Nombre": 'name',
    "Departamento": 'dpto',
    "Salario": 'salary',
    "Edad": 'age'
})
print(df_rename)
```

# Convertir una serie o columna de un tipo a otro => float => int

```python
df_rename['age'] = df_rename['age'].astype('Int64')
print(df_rename)
```

## Replace: Reemplazar valores dentro de una serie

```python
# reemplazar Ventas por ventas: 2 formas de hacerlo

#1ª forma:
# primero limpiar espacios y llevar a mayúsculas
df_rename['dpto'] = df_rename['dpto'].str.strip().str.upper() # pasa a mayúsculas todos los valores del campo dpto
print(df_rename)
print('#'*45)

#2ª forma: con replace
diccionario = {
    "Ventas": "COMERCIAL",
    "ventas": "COMERCIAL",
    "rrhh": "RRHH",
    "it": "IT"
}

df_rename['dpto'] = df_rename['dpto'].replace(diccionario) 

print(df_rename)

```

# PIPELINE
Es el conjunto de procesos que se realizan para la limpieza de datos, se realizan en el orden siguiente:

---
**Resumen — pipeline de limpieza típico:**
1. `df.isnull().sum()` → detectar nulos
2. `df.dropna()` o `df.fillna()` → tratar nulos
3. `df.duplicated()` → detectar duplicados
4. `df.drop_duplicates()` → eliminarlos
5. `df.rename()` / `.columns.str.lower()` → normalizar nombres
6. `.str.strip()` / `.replace()` → normalizar valores
