# Agregaciones y estadística en NumPy

Quizás, junto con operaciones, este sea uno de los temas más importantes de NumPy. Vamos a aprender a resumir conjuntos de datos, calcular media, moda, mediana, DT, máximo y mínimo, percentiles... Estos conceptos son los pilares básicos para entender un grupo de datos de un vistazo. 

````python
import numpy as np
````

## La media (promedio)
La media es el valor central o típio de un conjunto de datos. Se calcula sumando todos los valores del conjunto y dividiéndolo entre el número total de elementos. 

$$\bar{x} = \frac{x_1 + x_2 + \cdots + x_n}{n} = \frac{\sum_{i=1}^{n} x_i}{n}$$

**Ejemplo:** Las notas de un alumno son [5, 7, 8, 6, 9]. La media es:
$$\bar{x} = \frac{5 + 7 + 8 + 6 + 9}{5} = \frac{35}{5} = 7$$

````python
notas = np.array([5,6,2,7,1,8,9])
print('Media', np.mean(notas))
````

## Limitaciones de la media: los valores extremos la distorsionan. 

Imagina salarios de 5 personas. [1200, 1300, 1100, 1250, 50000]. El director cobra 50000 y los demás aprox. 1200€ de media. ¿Si hacemos la media qué pasa?

````python
salarios = [1200, 1300, 1100, 1250, 50000]
print('Media', np.mean(salarios))
#Salario medio 10970 no es representativo de la realidad. 
````

## La mediana

La mediana es el valor que queda justo en el centro cuando ordenamos los datos de menor a mayor. Es el valor por debajo del cual queda el 50% de los datos. 

- Número impar de datos: la mediana es el valor central
- Número par de datos: la mediana es la media de los dos valores centrales

**Ejemplo con 5 valores** (impar):
```
Datos: [3, 1, 7, 5, 2]  →  Ordenados: [1, 2, 3, 5, 7]
                                              ↑
                                        Mediana = 3
```

**Ejemplo con 6 valores** (par):
```
Datos: [4, 1, 7, 5, 2, 8]  →  Ordenados: [1, 2, 4, 5, 7, 8]
                                                  ↑↑
                                          Mediana = (4+5)/2 = 4.5
```

La mediana **no se deja engañar por valores extremos** (al contrario que la media). Es más representativa cuando existen valores extremos

````python
#usando salarios vamos a calcular la mediana
print('Media', np.mean(salarios))
print('Mediana', np.median(salarios))
#ordenación con NumPy
print('Salarios ordenados', np.sort(salarios,descending=False))
````

## La varianza y la desviación típica. 

La media y mediana nos dicen el valor central(son medidas de posición), pero no nos dicen **cuánto varían** los datos alrededor de ese valor. Para ello usamos la varianza y la desviación típica(son medidas de dispersión). 

** Ejemplo ** Dos clases de alumnos tiene la misma nota media 6, pero:
Clase A: [6,6,6,6,6] -> todos igual, sin variación
Clase B: [1,4,6,8,9] -> muy dispersa, gran variacion

## La varianza
Mide la dispersión media de los datos respecto de la media. Se calcula asi.

 - Calcular la media:  $\bar{x}$
 - Resta la media a cada dato: $(x_i - \bar{x})$
 - Elevamos al cuadrado para que no se cancelen los positivos y negativos: $(x_i - \bar{x})^2$
 - Calcular le media de los cuadrados

$$\text{Varianza} = \frac{\sum_{i=1}^{n}(x_i - \bar{x})^2}{n}$$

````python
clase_A = np.array([6,6,6,6,6])
clase_B = np.array([1,4,6,8,9])

print('varianza clase A', np.var(clase_A))
print('varianza clase B', np.var(clase_B))
````

## Desviación típica

Es simplemente la raíz cuadrada de la varianza. Se usa más que la varianza porque tiene las misma udes que los datos originales.
$$\sigma = \sqrt{\text{Varianza}}$$

````python
print('Desviación típica A', np.std(clase_A))
print('Desviación típica B', np.std(clase_B)) #cada nota se desvía 2.8 con respecto a la media
````

````python
print('### Clase_A ###')
print('Media', np.mean(clase_A))
print('Mediana', np.median(clase_A))
print('Varianza', np.var(clase_A))
print('Desviación típica', np.std(clase_A))
print('='*15)
print('### Clase_ ###')
print('Media', np.mean(clase_B))
print('Mediana', np.median(clase_B))
print('Varianza', np.var(clase_B))
print('Desviación típica', np.std(clase_B))
print('='*15)
````

````python
# desviación típica para este array => cómo lo haríamos sin la función std
datos = np.array([4,7,8,5,6])
media = np.mean(datos)

diferencias = datos - media
print(diferencias)

cuadrados = diferencias ** 2
print(cuadrados)

varianza = np.mean(cuadrados)
print(varianza)

desv_tipica = np.sqrt(varianza)
print(desv_tipica)
````

## Máximo, mínimo y rango

Son conceptos simples pero muy útiles

- **Mínimo**: El valor más pequeño del conjunto
- **Máximo**: El valor más grande del conjunto
- **Rango**: La diferencia entre el máximo y mínimo. Mide cuanto se extienden los datos.
  
$$\text{Rango} = \text{Máximo} - \text{Mínimo}$$

````python
temperaturas = np.array([15.2,18.5,22.1,25.8,28.3,26.0, 23.4, 19.7,16.1, 14.8])

print( 'Mínimo', np.min(temperaturas) )
print( 'Máxima', np.max(temperaturas) )
print( 'Rango', np.ptp(temperaturas), 'ºC (max-min)')
print( 'Media', np.mean(temperaturas))

print('Dia de la lista más frio', np.argmin(temperaturas), np.min(temperaturas))
print('Dia de la lista más frio', np.argmin(temperaturas), np.max(temperaturas))
````

## Percentiles y cuartiles

Un percentil (p) es el valor por debajo del cual queda el p% de los datos. 

**Ejemplo**: Si tu nota en un examen está en el percentil 80, significa que el 80% de los estudiantes ha sacado una nota igual o inferior a la tuya. Es decir, estás entre el 20% de las mejores notas de la clase. 

Los cuartiles dividen los datos en 4 partes iguales. 
```
Mínimo       Q1.      Q2.     Q3       Máximo
  | --------- | ------ | ----- |-------- |
        25%.      25%.     25%.    25%
```
El **rango intercuartílico** (IQR) =  Q3 - Q1
Mide la dispersión del 50% central de los datos. Es más robusto para el cálculo de la desviación típica por que ignora los extremos.


````python
notas_clase = np.array([3,4,4,5,5,5,6,6,6,6,7,7,7,7,7,8,8,8,8,8,8,9,9,10])

Q1 = np.percentile(notas_clase, 25)
Q2 = np.percentile(notas_clase, 50)
Q3 = np.percentile(notas_clase, 75)
IQR = Q3 - Q1
print('El 25% de los alumnos están por debajo de ', Q1)
print('El 50% de los alumnos están por debajo de ', Q2)
print('El 75% de los alumnos están por debajo de ', Q3)
print('El rango intercuartílico es ', IQR)

# ¿En qué percentil está un alumno con una nota de 8?
nota_alumno = 8
percentil = np.sum(notas_clase<=nota_alumno)/len(notas_clase)
centil = percentil*100 #centil 87
cuartil = percentil*4 #cuartil 3 #no estoy muy segura de esto, lo he hecho yo
decil = percentil *10 #decil 8  #igual, no muy segura
print(centil)
print(cuartil)
print(decil)
````

## Sumas acumuladas

La suma acumulada va sumando valores uno a uno, guardando el total de cada posición. Es útil para ver cómo crece un total a lo largo del tiempo(venta acumuladas, km recorridos, etc.) Equivalente a la tabla de frecuencias. 

````python
ventas_diarias = np.array([100, 150, 80, 200, 120, 90, 175])
dias = np.array(['Lun', 'Mar', 'Mié', 'Jue', 'Vie', 'Sáb', 'Dom'])
sum = np.sum(ventas_diarias)
acumulado = np.cumsum(ventas_diarias)#suma acumulada
print('Día      | Ventas  | Total acumulado')
print('-'*40)
for dia, v, ac in zip(dias,ventas_diarias,acumulado):
    print(f'{dia:6s}   |{v:6d}   |{ac:4d}  ') #6s son 6 espacios, 4d son 4 dígitos
````

# El parámetro axis: Manejar nuestro array n-dimensional por ejes. 
Podemos tratar solo las columnas/filas 

Con una tabla 2D (matriz) puedo querer:
- La media de cada columna
- La media de cada fila

El parámetro **axis** indica en qué dirección colapsan los datos:
    - axis = 1 recorre las columnas -> trabajo con las filas, opera hacia la derecha
    - axis = 0 recorre las filas -> trabajo con las columnas, opera hacia abajo

````python
notas = np.array([
    [6,7,8], #alumna Ana
    [4,5,7], #alumno Pablo
    [4,5,9], #alumna María
    [2,7,4]  #alumno Juan
])

media_asignatura = np.mean(notas, axis=0)
print('Mates', media_asignatura[0])
print('Lengua', media_asignatura[1])
print('Lengua', media_asignatura[2])


medias_alumnos = np.mean(notas, axis=1)
print('Media Ana', medias_alumnos[0])
print('Media Pablo', medias_alumnos[1])
print('Media María', medias_alumnos[2])
print('Media Juan', medias_alumnos[3])
````

**Resumen del tema:**

| Concepto | Fórmula | `numpy` | Para qué sirve |
|----------|---------|---------|----------------|
| Media | $\bar{x} = \frac{\sum x_i}{n}$ | `np.mean()` | Valor central típico |
| Mediana | valor del medio ordenado | `np.median()` | Centro robusto (no le afectan extremos) |
| Varianza | $\frac{\sum(x_i-\bar{x})^2}{n}$ | `np.var()` | Dispersión (en unidades²) |
| Desv. típica | $\sqrt{\text{Varianza}}$ | `np.std()` | Dispersión (en mismas unidades) |
| Mínimo/Máximo | — | `np.min()` / `np.max()` | Extremos del conjunto |
| Rango | max − min | `np.ptp()` | Amplitud total |
| Percentil P | valor bajo el que cae el P% | `np.percentile(a, P)` | Posición relativa |
| IQR | Q3 − Q1 | `np.percentile(a,75) - np.percentile(a,25)` | Dispersión central |
| Suma acumulada | — | `np.cumsum()` | Totales acumulados |
| `axis=0` | opera por columnas (↓) | `np.mean(a, axis=0)` | Resumir cada columna |
| `axis=1` | opera por filas (→) | `np.mean(a, axis=1)` | Resumir cada fila |
