## Contexto

Eres el profesor de matemáticas de 1º de Bachillerato.  
Tienes las **notas del examen de 25 alumnos** (escala 0–10) y quieres analizarlas con NumPy.

El array ya está preparado para que puedas empezar directamente:

```python
import numpy as np

np.random.seed(42)  # fija la aleatoriedad para obtener siempre los mismos valores
notas = np.round(np.random.uniform(2, 10, size=25), 1)
print("Notas de la clase:", notas)
```

---
## Ejercicio 1 — Conoce tu array

Muestra por pantalla:
- El número total de alumnos
- El tipo de dato del array
- El número de dimensiones

```python
print(f'El número total de alumnos es ',len(notas))# notas.size sería el código de NumPy
print(f'Tipo de dato del array: ',notas.dtype)
print(f'Dimensiones del array: ', notas.ndim)
```

---
## Ejercicio 2 — Slicing

Muestra las notas del **alumno 5 al alumno 10** (posiciones 4 a 9, recuerda que Python empieza en 0).

```python
notas_5_9 = notas[4:10]
print(notas_5_9)
```

---
## Ejercicio 3 — Nota más alta, más baja y media

Calcula con NumPy:
- La nota más alta
- La nota más baja
- La nota media de la clase

> Pista: usa `np.max()`, `np.min()` y `np.mean()`

```python
nota_mas_alta = np.max(notas)
nota_mas_baja = np.min(notas)
nota_media = np.mean(notas)
print(nota_mas_baja)
print(nota_mas_alta)
print(nota_media)
```

---
## Ejercicio 4 — Filtrado booleano: aprobados

Crea un array que contenga **solo las notas de los alumnos que han aprobado** (nota >= 5).  
Muestra también cuántos alumnos han aprobado.

```python
aprobados = notas>=5
resultado = notas[aprobados]
print(resultado)
#¿Cuántos han aprobado? 
cantidad = np.sum(aprobados)
print(f'{cantidad} de {len(notas)} alumnos han aprobados')


```

---
## Ejercicio 5 — Operación vectorial: cambio de escala

El director del centro quiere las notas en escala **0–100** (en lugar de 0–10).  
Convierte todo el array multiplicando cada nota por 10.  
Guarda el resultado en una variable llamada `notas_100`.

```python
notas_100 = np.int64(notas * 10)
print(notas_100)
```

---
## Ejercicio 6 — ¿Cuántos superan la media?

Usando el array original `notas` (escala 0–10), calcula cuántos alumnos tienen una nota **estrictamente superior a la media** de la clase.

```python
alumnos_superior_media = np.sum(notas > nota_media)
print(alumnos_superior_media)
```

---
## Ejercicio 7 — Suspensos

Crea un array `suspensos` con las notas de los alumnos que NO han aprobado (nota < 5).  
Muestra cuántos son y sus notas.

```python
suspensos = notas < 5
resultado = notas[suspensos]
print(resultado)
print(np.sum(suspensos))
```

---
## Ejercicio 8 — np.where: etiquetas

Usa `np.where()` para crear un array de texto donde:
- Si nota >= 5 → `"Aprobado"`
- Si nota < 5 → `"Suspenso"`

Muestra el resultado junto a las notas originales.

```python
aprobado_suspenso = np.where (notas >=5, 'Aprobado', 'Suspenso')
print(aprobado_suspenso)
print(' ')
for n, a in zip(notas, aprobado_suspenso):
    print(f'Nota: {n} -> {a}' )
```
