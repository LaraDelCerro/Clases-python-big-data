## Operaciones con NumPy

En este tema vamos aprender como operar matematicamente sobre arrays sin tener que recorrelos. 
Sin bucles. Esto es uno de los conceptos mas importes de NumPy

```python
import numpy as np
```

```python
examen_mates = [3.5, 5, 6.5, 2, 10, 9]
examen_lengua = [6.5, 1, 3, 4, 7, 8]

examen_mat_np = np.array(examen_mates)
examen_lng_np = np.array(examen_lengua)

print((examen_mat_np + examen_lng_np)/2) #nos daría la nota media de cada alumno 
```

```python
a = np.array([10,20,30,40,50,60])
b = np.array([2,4,6,8,10,12])

print('a+b: ', a + b)
print('a-b: ', a -b )
print('a*b: ', a*b)
print('a/b: ', a/b)
print('a**2: ', a**2)
print('b % 3: ', b%3) #el módulo, resto de la división
```

## Broadcasting: cuando operamos con elementos de distitna longitud

```python
array_precio_productos = np.array([ [20, 15, 60],
                                  [30, 45, 39.90],
                                  [10, 12.90, 30],
                                  [20, 15.30, 60]])

lista_precios_envios = np.array([12, 3, 8.9])

print(array_precio_productos + lista_precios_envios)
```

## Operaciones de comparacion

Las comparaciones (`>`, `<`, `==`, `!=`, `>=`, `<=`) también se aplican elemento a elemento y devuelven un array de `True`/`False`:

```python
notas  = np.array([4, 7, 6, 3, 5, 10, 8, 7, 9])
print(notas)
#¿Quién aprobaría?
aprobados = notas >=5
print(aprobados)

#preguntas globales
print('¿Algún alumno suspende?', np.any(notas<5)) #True
print('¿Todos aprueban?', np.all(notas>=5)) #False
print('¿Cuántos alumnos aprueban?', np.sum(aprobados)) #True=1, False=0
print('¿Cuántos alumnos suspenden?', np.sum(notas<5))
```

## Funciones especialeas => ufuncs

Se trata de operaciones como  raíz cuadrada, exponenciales, logarítmicas, redondeos, valores absolutos, senos, cosenos

```python
numeros = np.array([-2, 1.3, 0.8, -5.3, 1.0, 7, 29])

# valor absoluto
numeros_abs = np.abs(numeros)
print('valor absoluto', numeros_abs)

#exponencial - b^e: Crecimiento de poblaciones, interés compuesto, magnitudes...
#si inviertes 1000€ al 5% continuo, tras tres años tienes: 1000 * e
anyos = np.array([0,1,5,10,20,40])
capital = np.round(1000*np.exp(0.02*anyos),2)
print('exponencial', capital)


#logaritmo, es la inversa de la exponencial. Por ejemplo, se usa para calcular la escala Ritchter
#Escala de Rithcer cada grado = 10 x energía
#magnitud = log10(energia)
energias = np.array([10, 100, 1000, 10000, 100000, 1000000])
magnitudes = np.log10(energias)
print(magnitudes)
#representación gráfica de magnitudes y energías
for e, m  in zip(energias, magnitudes):
    print(f"Energia: {e} -> Magnitud: {m}")

# Raíz cuadrada
resultado = np.sqrt([4, 9, 16, 25, 36, 49, 64, 81])

# Redondeos
numeros_racionales = np.array([2.456, 3.6543, 67.234,9.2345,10.345635])
#Redondeo matemático
print(np.round(numeros_racionales, 2))
#Redondeo a la baja
print(np.floor(numeros_racionales))
#Redondeo a la alta
print(np.ceil(numeros_racionales))
#Trunc- trunca el nº eliminando parte decimal, siempre hacia cero, equivalente a floor
print(np.trunc(numeros_racionales))
```

# np.where condicional vectorizado

```python
notas = np.array([4.5, 7.2, 3.1, 8.8, 5.0, 6.3])

print(np.where(notas >= 5, "Aprobado", "Suspenso"))
```
